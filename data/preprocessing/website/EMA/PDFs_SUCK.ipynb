```python
def headers_lines_main(pdf_main):

    lines_tagged_lst = []
    
    curr_chapter = None
    curr_subheader1 = None
    curr_subheader2 = None


    for page in pdf_main:
        blocks = page.get_text("dict")["blocks"]
        for b in blocks:  # iterate through the text blocks
            if b['type'] == 0:  # this block contains text
                for l in b["lines"]:  # iterate through the text lines
                    for s in l["spans"]:  # iterate through the text spans
                        text = s["text"].strip().replace("�", " ")
                        if s["size"] > 10: # update current chapter if font size is large
                            curr_chapter = text
                            curr_subheader2 = None # reset subheader2
                        elif s["font"] == "Univers-BoldOblique": # update subheader 1 if italic
                            curr_subheader1 = text
                            curr_subheader2 = None # reset subheader2
                        elif (s["font"] == 'Univers-Light') & (s["color"] != 0): # update subheader 2 if same font as regular text but d
                            curr_subheader2 = text

                        lines_tagged_lst.append((curr_chapter, curr_subheader1, curr_subheader2, text))

    return lines_tagged_lst
```

```python
pdf = fitz.open("C:/Users/Zhong Xuean/Documents/dsaid-hackathon23-illuminati/data/preprocessing/raw/Handbook_for_Solar_PV.pdf")
pdf_main = pdf.pages(5, 31)
```

```python
headers_lines_m = headers_lines_main(pdf_main)
```
